### Team Management System


```python
class kbfc:
    
    def __init__(self,listofplayers):
        self.availableplayers=listofplayers

    def displayavailableplayers(self):
        print("All players in the team are:")
        print("-----------------------------")
        for players in self.availableplayers:
            print(players)
    def removeplayer(self,requestedplayer):
        if requestedplayer in self.availableplayers:
            self.availableplayers.remove(requestedplayer)
            print("Contract of this player is terminated")
        else:
            print("Sorry the player that u want to end contract , is not in this team")

    def addplayer(self,returnplayer):
        self.availableplayers.append(returnplayer)
        print("New player added to the team .")



class Coach:

    def delplayer(self):
        print("Enter the name of the player you'd like to end contract:")
        print("------------------------------")
        self.players=input()
        print("------------------------------")
        return self.players

    def newplayer(self):
        print("Enter the name of the player you'd like to sign:")
        print("------------------------------")
        self.player=input()
        print("------------------------------")
        return self.player 


def call():
    team=kbfc(["KARANJIT SINGH","PRABHSUKHAN GILL","HARMANJOT KHABRA","JESSEL CARNEIRO","MARCO LESCOVIC","RUIVAH HORMIPAM","ADRIAN LUNA","AYUSH ADHIKARI","JEAKSON SINGH","SAHAL ABDUL SAMAD","KP RAHUL"]) 
    coach=Coach() 
    done=False 
    while done==False:
        print(""" 
        WELCOME TO KERALA BLASTERS FC
        
          1. Display all  players in the team
          2. Sign a new player
          3. End contract of a player
          4. Exit
        
          """)
       
        try:
            choice=int(input("Enter Choice:"))
            print("------------------------------")
            if choice==1:
                team.displayavailableplayers()
            elif choice==2:
                team.addplayer(coach.newplayer())
            elif choice==3:
                team.removeplayer(coach.delplayer())
            elif choice==4:
                print("Thank you !")
                break 
                            
        except:
            print("Please enter proper choice.")

call()
```
